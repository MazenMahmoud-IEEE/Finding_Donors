# Importing Libraries

```python
import pandas as pd
import numpy as np
from IPython.display import display
import seaborn as sns
import matplotlib.pylab as plt
import warnings
warnings.filterwarnings('ignore')
```

# Reading and copying the dataset

```python
donors = pd.read_csv("census.csv")
print(donors.shape)
display(donors)
donors_cpy = donors.copy()
```

# Removing Duplicates

```python
donors_cpy = donors_cpy.drop_duplicates()
display(donors_cpy)
```

# Dealing with Nulls

```python
nulls = donors_cpy.isnull().sum().sort_values(ascending=False)
print(nulls)
```

# Data Preprocessing

```python
donors_cpy.info()
```

```python
donors_cpy['income'] = donors_cpy['income'].map({'<=50K': 0, '>50K': 1})
donors_cpy
```

```python
donors_cpy.info()
```

```python
donors_cpy.select_dtypes(include='object').nunique().index
```

```python
donors_cpy['sex'] = donors_cpy['sex'].str.strip().map({'Female': 0, 'Male': 1})
donors_cpy
```

# Multicategory columns

```python
# List of columns with more than two unique values
multi_category_columns = ['workclass', 'education_level', 'marital-status', 'occupation',
       'relationship', 'race', 'native-country']

# Apply one-hot encoding
features = pd.get_dummies(donors_cpy, columns=multi_category_columns, drop_first=True)  # drop_first=True to avoid dummy variable trap
#print(features.to_string())
display(features)
```

# Converting boolean columns (True & False) into int (0 & 1)

```python
# Convert boolean values to 0 and 1
features = features.astype(float)

# Display the updated DataFrame
display(features)
```

# Correlation

```python
corr = features.corr()['income']
corr
```

```python
top_features = corr[(corr > 0.1) | (corr < - 0.1)].sort_values(ascending=False) #displays  correlation coefficient greater than 0.5
top_features = top_features.drop('income')
top_features.index
```

# Features and Target

```python
target = features['income']
target
```

```python
features.drop('income' ,axis=1 , inplace=True)
features = features[top_features.index]
features
```

# Splitting the data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)
```

```python
target.value_counts() 
# There is class imbalance in the data
```

# Testing all base models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score
from sklearn.model_selection import ShuffleSplit
from imblearn.over_sampling import SMOTE

## building different models
def model_building(model_name):
  model = model_name
  model.fit(X_train, y_train)
  print(f"******** Model :- {model_name} ********\n")
  y_test_pred = model.predict(X_test)
  y_train_pred = model.predict(X_train)
  train_f1 = f1_score(y_train, y_train_pred)
  test_f1 = f1_score(y_test, y_test_pred)
  train_accuracy = accuracy_score(y_train, y_train_pred)
  test_accuracy = accuracy_score(y_test, y_test_pred)
  print(f"Training F1 Score: {train_f1:.4f}, Training Accuracy: {train_accuracy:.4f}")
  print(f"Testing F1 Score: {test_f1:.4f}, Testing Accuracy: {test_accuracy:.4f}")
  print()

## dictionary with different models
model_dict = {'dt':DecisionTreeClassifier(criterion='entropy',random_state=0,class_weight='balanced'), 
              'knn':KNeighborsClassifier(n_neighbors=17), 
              'rf': RandomForestClassifier(random_state=0,class_weight='balanced'), 
              'xgb':xgb.XGBClassifier(random_state=0, class_weight='balanced')
              #,'svm':SVC(random_state=0,class_weight='balanced')
}


## calling to build and evaluate models
for key in model_dict.keys():
  model_building(model_dict[key])
```

# Found best base model : Gradient Boost

# Tuning Gradient Boost Model

## BayesianSearch

```python
from skopt import BayesSearchCV
from xgboost import XGBClassifier

# Define the hyperparameter search space
param_space = {
    'learning_rate': (0.1, 1.0, 'log-uniform'),
    'max_depth': (2, 10),
    'n_estimators': (100, 500),
    'subsample': (0.5, 1.0),
    'colsample_bytree': (0.5, 1.0)
}

# Set up the XGBoost model and Bayesian search
xgb_model = XGBClassifier()
bayes_search = BayesSearchCV(estimator=xgb_model, search_spaces=param_space, n_iter=30, cv=5,scoring='f1')

# Fit the model to find the best hyperparameters
bayes_search.fit(X_train, y_train)
best_model = bayes_search.best_estimator_
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

# Calculate F1 score and accuracy for training and testing sets
train_f1 = f1_score(y_train, y_train_pred)
test_f1 = f1_score(y_test, y_test_pred)
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# Output the best parameters found
print(bayes_search.best_params_)
print(f"Training F1 Score: {train_f1:.4f}, Training Accuracy: {train_accuracy:.4f}")
print(f"Testing F1 Score: {test_f1:.4f}, Testing Accuracy: {test_accuracy:.4f}")
```
